```python
import cv2 as cv
```

```python
# haarcascade pretrained obeject detected algorithem 
face_cascade=cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face_cascade
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv.imread('th.jpeg')
plt.imshow(img)
```

```python
image=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# testing the face is detected or not using face cascade
faces=face_cascade.detectMultiScale(image ,1.1 ,4 ,minSize=(30,30))
```

```python
faces
```

```python
# CascadeClassifier face_cascade
```

```python
len(faces)
```

```python

# plot the face
# x,y -cordinates ,w-width,h -height
for (x,y,w,h) in faces:
    cv.rectangle(image,(x-5,y-5),(x+w+5 , y+h+5),(0,0,0),2)
```

```python
plt.imshow(image)
```
