```python

grey=cv.imread('Mammootty.jpg',cv.IMREAD_GRAYSCALE)
cv.imshow("image",grey)
cv.waitKey()
cv.destroyAllWindows()
/ -- convert it into gray scale
```

```python

(thresh,bw)=cv.threshold(grey,0,255,cv.THRESH_BINARY | cv.THRESH_OTSU)
cv.imshow("image",bw)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# edge ditection
edges=cv.Canny(grey,150,150)
cv.imshow("image",edges)

cv.waitKey()
cv.destroyAllWindows()
```
